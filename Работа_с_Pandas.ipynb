# Лабораторная работа 2
1. Посчитать площадь и объём товаров. 
2. Посчитать среднее в категории, если у товаров нет данных по высоте, ширине и длине.
3. Исключить товары у которых частота продаж меньше 0.012 и у которых нет остатков и транзита
4. Отсортировать список по частоте продаж товаров. 

## Посчитать площадь и объём товаров

```python
data["Объём"] = data["ДлинаЕд"] * data["ШиринаЕд"] * data["ВысотаЕд"]
data["Площадь"] = 2 * (data["ДлинаЕд"] *  data["ШиринаЕд"] + data["ДлинаЕд"] * data["ВысотаЕд"] + data["ШиринаЕд"] * data["ВысотаЕд"])
data
```

## Посчитать среднее в категории, если у товаров нет данных по высоте, ширине и длине

```python
def check(df):
  return df.loc[(df["ДлинаЕд"] != 0.0) & (df["ШиринаЕд"] != 0.0) & (df["ВысотаЕд"] != 0.0)]

df_mean = check(data).groupby("КодКатегории").mean()
df_mean
```

```python
for index, row in df_mean.iterrows():
  data.loc[(data["Объём"] == 0.0) & (data["КодКатегории"] == index), "Объём"] = row["Объём"]
  data.loc[(data["Площадь"] == 0.0) & (data["КодКатегории"] == index), "Площадь"] = row["Площадь"]

data
```

## Исключить товары у которых частота продаж меньше 0.012 и у которых нет остатков и транзита

```python
data = data.loc[(data["ЧастотаПродаж"] >= 0.012) & ((data["ВсегоТовара"] > 0.0) | (data["Транзит"] > 0.0))]
data
```

# Лабораторная работа 3
1. Каждый товар имеет свою занимаемую площадь. Нужно рассчитать сколько товаров данных категорий влезет по площади. Для этого нужно сгенерировать для каждой уникальной категории площадь и рассчитать.

```python
names = ['ТВ', 'Компы','Комплектующие', 'Телефоны']
categories = pd.DataFrame()

for elem in names:
  df = pd.read_excel('/content/drive/MyDrive/СЯП/Товары.xlsx', sheet_name=elem, names=['КодКатегории', 'ПлощадьКатегории'])
  df['Категория'] = elem
  df['ПлощадьКатегории'] = df.iloc[0]['ПлощадьКатегории']
  df['ОстатокОтПлощади'] = df.iloc[0]['ПлощадьКатегории']
  categories = pd.concat([categories, df])

categories
```

```python
def count_the_area(df, name, area):
  for index, row in df.iterrows():
    if row['Категория'] == name:
      area -= row['Площадь']
    else:
      name = row['Категория']
      area = row['ПлощадьКатегории']
    df.loc[index, 'ОстатокОтПлощади'] = area
  return df.loc[df['ОстатокОтПлощади'] >= 0.0]

df = count_the_area(data.merge(categories, on='КодКатегории', how="left").sort_values(by='Категория'), 
                    df.iloc[0]['Категория'], 
                    df.iloc[0]['ПлощадьКатегории'])
df
```
