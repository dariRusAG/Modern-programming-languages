## Выполнение заданий

```python
data_del = data.drop(['ДлинаЕд.1','ШиринаЕд.1','ВысотаЕд.1','ВсегоТовара','НаВитрине','Транзит.1'], axis=1)
data_del
```

```python
data_rename = data_del.rename(columns={"Остаток": "ВсегоТовара", "Количество": "НаВитрине"})
data_rename
```

```python
data_rename.groupby("КодКатегории").mean()['Цена'].reset_index()
```

```python
data_dropna = data_rename.fillna(0)
data_dropna
```

```python
data_dropna["РазницаЦены"] = data_dropna['Цена'] - data_dropna.groupby(["КодКатегории"])['Цена'].transform('mean')
data_dropna
```
